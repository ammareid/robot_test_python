```python
import sys
import numpy as np
from rplidar import RPLidar


PORT_NAME = 'COM3'


def run(path):
    '''Main function'''
    lidar = RPLidar(PORT_NAME)
    data = []
    try:
        print('Recording measurments... Press Crl+C to stop.')
        for scan in lidar.iter_scans():
            data.append(np.array(scan))
    except KeyboardInterrupt:
        print('Stoping.')
    lidar.stop()
    lidar.disconnect()
    np.save(path, np.array(data))

if __name__ == '__main__':
    run(sys.argv[1])
```

```python
import time
from datetime import datetime
from rplidar import RPLidar

lidar = RPLidar('COM3',baudrate=115200)

info = lidar.get_info()
print(info)

health = lidar.get_health()
print(health)

for scanNo, scan in enumerate(lidar.iter_scans()):
#     print('%d: Got %d measurments' % (i, len(scan)))
    closestAngle = None
    closestDistance = None

    for lineNo, line in enumerate(scan):
        angle = line[1]
        distance = line[2]
        if(lineNo==0):
            closestAngle = angle
            closestDistance = distance
        else:
            if(distance<closestDistance):
                closestAngle = angle
                closestDistance = distance
#         print("Angle: ", angle, "Distance: ", distance)
    print("Scan No: ", scanNo, ", Time: ", datetime.now(), ", Closest Angle: ", closestAngle, ", Closest Distance: ", closestDistance/10)
    if scanNo >= 50:
        break

# scan = lidar.iter_scans()

lidar.stop()
lidar.stop_motor()
lidar.disconnect()
```
